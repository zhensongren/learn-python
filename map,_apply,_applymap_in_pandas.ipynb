```python
import pandas as pd
import numpy as np
df = pd.DataFrame({ 'A': [1,2,3,4], 
                   'B': [10,20,30,40],
                   'C': [20,40,60,80]
                  }, 
                  index=['Row 1', 'Row 2', 'Row 3', 'Row 4'])
df
```

```python
df.info()
```

## apply

```python
def get_sum(row):
    return row.sum()
    
df.loc['Row 5'] = df.apply(get_sum, axis=0)
df
```

apply a function along the row(axis = 1) or column(axis = 0). 

```python
df['D'] = df.apply(np.sum, axis=1)
df
```

```python
df['even_A'] = df['A'].apply(lambda x:2*x if x%2!=0 else x)
df
```

## map

```python
df['E'] = df['A'].map(np.square)
df
```

```python
df['odd_A'] = df['A'].apply(lambda x:2*x + 1 if x%2==0 else x)
df
```

# applymap

Python function, returns a single value from a single value.

```python
df.applymap(lambda x:0 if x < 5 else x)
```

```python
df.applymap(lambda x:0)
```
