# using Recursion to generatre a list of n Fibonacci numbers. F_n =  F_n-1 + F_n-2


```python
def FibLst(n): # n is the number of element in the list
    if n <= 0: return -1
    nums = []
    # get the nth F number, O(n)
    for i in range(1, n+1):
        nums.append(Fib(i)) # O(n)
    return nums

def Fib(n):
    if n == 1: return 0
    if n == 2: return 1
    return Fib(n-1) + Fib(n-2)    

%timeit FibLst(10)
FibLst(10)
```

# iterative from 1 to n

```python
def Fib1ton(n):
    if n < 1: return "incorrect input"
    if n == 1: return [0]
    if n == 2: return [0, 1] 
    
    nums = [0, 1]
    for i in range(2, n):
        nums.append(nums[i-1] + nums[i-2])
        
    return nums

%timeit Fib1ton(10)
```

```python
def reverse2(list):
    i, j = 0, len(list)-1
    while i>j:
        list[i], list[j] = list[j], list[i]
        i+=1
        j-=1
    return list
reverse2(nums)
```

```python
def RmDuplicate2(nums):
    if type(nums[0])==list: 
        nums = nums[0]

    nums.sort()
    res = [nums[0]]
    for num in nums:
        if num != res[-1]:
            res.append(num)
    return res

RmDuplicate2(nums)
```

```python
%timeit RmDuplicate2(nums)
```

```python
def num_insert(nums, num):
    # find the index at which location to insert and move the rest of elements on index to the end
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = low + (high - low) // 2
        if nums[mid] == num:
            nums.insert(mid, num)
        elif num > nums[mid]:
            low = mid + 1
        else: high = mid - 1
            
    nums.insert(low, num)
    return nums
```

```python
# use binary search to find the index to insert and create a new list by copying the old and newly added num
def NumInsert(nums, num):
    if num < nums[0]: 
        return [num] + nums
    if num > nums[-1]:
        return nums + [num]
    # find the index for the new element
    left, right = 0, len(nums)-1
    while left<=right:
        mid = left + (right - left)//2
        if nums[mid]==num:
            return nums[:mid+1] + [num] + nums[mid+1:]
        
        elif num>nums[mid]:
            left = mid + 1
        else:
            right = mid - 1 
    return nums[:left+1] + [num] + nums[left+1:]
```

```python
%timeit NumInsert(nums, num)
```

```python
def sort_rm(nums):
    res = []
    for num in nums:
        if num not in res:
            res.append(num)
            
    return res

%timeit sort_rm([1,2,2,3,4,4,5,6])
```

```python
def merge_sorted(nums1, nums2):
    res = []
    i, j = 0, 0
    # Copy over the 2 list to a 3rd list.
    while i < len(nums1) - 1 and j < len(nums2) - 1:
        if nums1[i] < nums2[j]:
            res.append(nums1[i])
            i += 1
            
        else: 
            res.append(nums2[j])
            j += 1
            
    # either i or j is at the end of array index. Copy over the rest of the other list.
    if i == len(nums1) - 1:     
        res += nums2[j:]
    if j == len(nums2) - 1:
        res += nums1[i:]
        
    return res

%timeit merge_sorted([1,2,3,4], [2,4,5,7])
```
