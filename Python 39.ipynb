```python
# a list of odd numbers between i and j inclusive
def odd_nums(i, j):
    if i>j: return
    return [x for x in range(i, j+1) if x%2!=0]

%timeit odd_nums(1, 5)
odd_nums(1, 5)
```

```python
def c2f(c):
    f = [x*1.8+32 for x in c]
    return f

c2f([1,2,3,4,100])
```

# using Recursion to generatre a list of n Fibonacci numbers. F_n =  F_n-1 + F_n-2


```python
def FibLst(n): # n is the number of element in the list
    if n <= 0: return -1
    nums = []
    # get the nth F number, O(n)
    for i in range(1, n+1):
        nums.append(Fib(i)) # O(n)
    return nums

def Fib(n):
    if n == 1: return 0
    if n == 2: return 1
    return Fib(n-1) + Fib(n-2)    

%timeit FibLst(10)
FibLst(10)
```

```python
def Fib(n):
    if n<=0:
        print("Incorrect input")
    # First Fibonacci number is 0
    elif n==1:
        return 0
    # Second Fibonacci number is 1
    elif n==2:
        return 1
    else:
        return Fib(n-1)+Fib(n-2)
      
def FibLst(n:int):
    if n < 0: return -1
    f = []
    for i in range(1, n+1):
        f.append(Fib(i))
    return f   
```

# iterative from 1 to n

```python
def Fib1ton(n):
    if n < 1: return "incorrect input"
    if n == 1: return [0]
    if n == 2: return [0, 1] 
    
    nums = [0, 1]
    for i in range(2, n):
        nums.append(nums[i-1] + nums[i-2])
        
    return nums

%timeit Fib1ton(10)
```
